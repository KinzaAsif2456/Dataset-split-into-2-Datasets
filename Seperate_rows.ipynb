```python
import numpy as np
import pandas as pd
```

```python
df= pd.read_csv("annual_temp.csv")
print("The Data of Annual Temperature: \n" ,df)
```

```python
# Separate the rows by source using groupby and dictionary comprehension for efficiency
"""use the groupby method to group the data by the 'Source' column and create a dictionary data_by_source
where each key is a source (GCAG or GISTEMP) and each value is the corresponding DataFrame."""
data_by_source = {source: data for source, data in df.groupby('Source')}


# Save the separated data to new CSV files in the /content/ directory
data_by_source['GCAG'].to_csv("/content/annual_temp_gcag.csv", index=False)
data_by_source['GISTEMP'].to_csv("/content/annual_temp_gistemp.csv", index=False)

print("GCAG data:")
print(data_by_source['GCAG'])
print("\nGISTEMP data:")
print(data_by_source['GISTEMP'])
```
